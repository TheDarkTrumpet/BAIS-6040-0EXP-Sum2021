# Printing very basic mathematical functions

The easiest demonstration is to plot a simple exponential function

```python
import numpy as np

# Define our x range, -1 to 99
x = np.arange(-1, 100)

# Define our y function, exponential is x ** 2
y = x ** 2

print(f"First 10 elements of x => {x[:10]}")
print(f"First 10 elements of y => {y[:10]}")
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, y)
# plt.show()
```

# Printing "layers"
The way that Matplotlib works is that it's a layer-based diagramming tool.  Meaning each call
to an existing "buffer" sits on top of the previous one.  But, this is cleared after a cell.  Look at the following two examples.

```python
y1 = x
print(f"First 10 elements of x => {x[:10]}")
print(f"First 10 elements of y => {y1[:10]}")
```

```python
# Note the graph here.  It's smaller, it also only has one graph on it.
plt.plot(x, y1, linestyle='dashed', color='blue')
plt.show()
```

```python
# To get this all working together, given the way we approached plots
# need to be done in one cell, the plt.show() is largely voluntary
# and avoids printing object information.
x = np.arange(-1, 100)
y = x ** 2
y1 = x
plt.figure(figsize=(10,10))
plt.plot(x, y)
plt.plot(x, y1, linestyle='dashed', color='blue')
plt.show()
```

```python
# An alternate method is through some assignment
exponentialFigure = plt.figure(figsize=(10,10))
exponentialFigure.show()
plt.plot(x, y)
```

```python
plt.plot(x, y1, linestyle='dashed', color='blue')
```

# Other types of plots

There are many types of plots that matplotlib can do, and I won't be covering them all.  The best one can do is read the documentation and spend time trying other ones.  But, I'm including some examples below.  They're taken from "Python Data Science Handbook"

## Histogram

```python
np.random.seed(1000)
data = np.random.randn(1000)
print(f"First 10 elements {data[:10]}")
```

```python
plt.hist(data, bins=30,  histtype='stepfilled', color='steelblue')
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

# Two-dimensional Histograms
This section was taken from "Python Data Science Handbook", page 247-249

Other documentation to help understand what's going on:
* https://numpy.org/doc/stable/reference/random/generated/numpy.random.multivariate_normal.html
* https://numpy.org/doc/stable/reference/generated/numpy.ndarray.T.html

```python
mean = [0,0]
cov = [[1, 1], [1,2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

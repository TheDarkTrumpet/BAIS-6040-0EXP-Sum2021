# 8.02.01 - Seaborn

## Purpose

Seaborn sits on top of Matplotlib, and has some nice extra features.  Part of it is that it's a bit nicer to look at.  Second, it can work with Pandas a lot nicer.

## Libraries
* Pandas
* Matplotlib
* Seaborn


```python
# Basic setup/example, from their website

# Import seaborn
import seaborn as sns
%matplotlib inline
# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
import pandas as pd
pieRates = pd.read_csv("../../data/pie_rates.csv")
pieRates
```

```python
sns.relplot(
    data=pieRates, x="Pie", y="Rate", col="Country",
    size="Rate"
)
```

```python
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
barchart, axis = plt.subplots(figsize=(10,6))
axis = sns.barplot(x="Pie", y="Rate", hue="Country", data=pieRates, ci=None, palette="muted",orient='v')
axis.set_title("Pie Rates!")
axis.set_xlabel("Country")
axis.set_ylabel("Rate")
```

# Using as functions

```python
def generateImage(exp = 2, color='Blue'):
    plt.figure(figsize=(10, 10))
    xRange = np.arange(-1, 100)
    yRange = x ** exp
    plt.plot(xRange, yRange, color=color)
```

```python
generateImage()
generateImage(3, 'Purple')
```
